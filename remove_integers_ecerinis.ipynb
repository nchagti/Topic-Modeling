```python
import re

def remove_integers_from_uploaded_file(input_path, output_path):
    """Remove all integers from a text file and save the cleaned version."""
    with open(input_path, 'r', encoding='utf-8') as file:
        content = file.read()  # Read the file content
    
    cleaned_content = re.sub(r'\d+', '', content)  # Remove all sequences of digits\

    cleaned_lines = [line.lstrip() for line in cleaned_content.splitlines()]
    cleaned_content = '\n'.join(cleaned_lines)
    
    with open(output_path, 'w', encoding='utf-8') as file:
        file.write(cleaned_content)  # Write cleaned content to output

# Paths for input and output
input_path = "Ecerinis_spoken_text.txt"  # The uploaded file
output_path = "Ecerinis_spoken_text.txt"  # New file for cleaned text

# Apply the function
remove_integers_from_uploaded_file(input_path, output_path)
```
