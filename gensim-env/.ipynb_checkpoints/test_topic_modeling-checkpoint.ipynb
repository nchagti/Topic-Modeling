### Topic Modeling with Gensim

```python
from gensim import corpora
from gensim.models import LdaModel, CoherenceModel, LsiModel, HdpModel
#pyLDAvis
#import sys
#!{sys.executable} -m pip install pyLDAvis

import json
```

```python
# Load the processed corpus
import pickle

# Load the processed corpus from the .pkl file
with open("../processed_corpus.pkl", "rb") as f:
    processed_corpus = pickle.load(f)

# Now you can access tokens directly
tokens_1 = processed_corpus[0]["tokens"]
print(type(tokens_1))  # Should be <class 'list'>
```

```python
print(type(processed_corpus))
```

```python
# Extract token lists from processed_corpus
token_lists = [tragedy["tokens"] for tragedy in processed_corpus]
```

```python
for i in range(11):
    print(len(token_lists[i])) #check length of tokens for each play
```

```python
print(processed_corpus[0]['title']) #get title for play
```

```python
# Create a dictionary from the tokenized texts
dictionary = corpora.Dictionary(token_lists)

# Filter out extreme tokens (optional)
dictionary.filter_extremes(no_below=2, no_above=0.5) #no_below=2 removes words that appear in fewer than 2 plays #no_above=0.5 removes words that appear in more than 50% of the plays

# Create a bag-of-words representation of the corpus
corpus = [dictionary.doc2bow(tokens) for tokens in token_lists]
```

```python
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

#### LDA — Latent Dirichlet Allocation
In this, observations (e.g., words) are collected into documents, and each word's presence is attributable to one of the document's topics. Each document will contain a small number of topics.

```python
with open("corpus.pkl", "rb") as f:
    corpus = pickle.load(f)
dictionary = corpora.Dictionary.load('dictionary.gensim')
```

```python
lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=10
)
```

```python
lda_model.show_topics()
```

```python
for idx, topic in lda_model.print_topics(num_words=10):
    print(f"Topic #{idx + 1}: {topic}")
```

#### LSI — Latent Semantic Indexing 
LSI, also known as Latent Semantic Analysis (LSA), is a technique that uncovers latent topics by decomposing the term-document matrix using Singular Value Decomposition (SVD).

```python
lsi_model = LsiModel(corpus=corpus, num_topics=10, id2word=dictionary)
lsi_model.show_topics(num_topics=5)
```

#### HDP — Hierarchical Dirichlet Process
HDP is a nonparametric Bayesian approach to topic modeling. Unlike LSI and LDA, HDP can automatically determine the number of topics.

```python
hdp_model = HdpModel(corpus=corpus, id2word=dictionary)
hdp_model.show_topics()[:5]
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
```

```python
for i, bow in enumerate(corpus):
    dist = lda_model.get_document_topics(bow, minimum_probability=0.0)
    print(f"\n{processed_corpus[i]['title']}")
    for topic_id, weight in sorted(dist, key=lambda x: -x[1]):
        print(f"  Topic #{topic_id + 1}: {weight:.3f}")
```

```python
topic_id = 9  # Gensim topic index that corresponds to pyLDAvis Topic 2

results = []
for i, bow in enumerate(corpus):
    dist = dict(lda_model.get_document_topics(bow, minimum_probability=0.0))
    weight = dist.get(topic_id, 0)
    results.append((processed_corpus[i]["title"], weight))

# Sort by weight descending
results = sorted(results, key=lambda x: -x[1])
for title, weight in results:
    print(f"{title:<25} → Topic 2 weight: {weight:.3f}")
```

```python
document_topics = []
for i, bow in enumerate(corpus):
    topic_dist = lda_model.get_document_topics(bow)
    document_topics.append((processed_corpus[i]["title"], topic_dist))
```

```python
for title, topics in document_topics:
    # Get topic with highest probability
    top_topic = max(topics, key=lambda x: x[1])
    print(f"{title:<20} → Topic #{top_topic[0] + 1} (weight: {top_topic[1]:.3f})")
```

```python
#get topic distribution for each play
for title, topics in document_topics:
    print(f"\n{title}")
    for topic_id, weight in sorted(topics, key=lambda x: -x[1]):
        print(f"  Topic #{topic_id + 1}: {weight:.3f}")
```

```python
import pandas as pd

rows = []
for i, (title, topics) in enumerate(document_topics):
    topic_weights = [0] * lda_model.num_topics
    for topic_id, weight in topics:
        topic_weights[topic_id] = weight
    rows.append([title] + topic_weights)

df_topics = pd.DataFrame(rows, columns=["Title"] + [f"Topic {i+1}" for i in range(lda_model.num_topics)])
```

### Attempting to retrain LDA model by filtering out ultra-common words

```python
#get most common tokens across the corpus
from collections import Counter

# Flatten all tokens into one list
all_tokens = [token for doc in processed_corpus for token in doc["tokens"]]
token_counts = Counter(all_tokens)
ultra_common_auto = set([word for word, _ in token_counts.most_common(10)])
```

```python
print(ultra_common_auto)
```

```python
#get most common tokens in each document

from collections import defaultdict

# Count how many documents each word appears in
doc_freq = defaultdict(int)

for doc in processed_corpus:
    unique_tokens = set(doc["tokens"])
    for token in unique_tokens:
        doc_freq[token] += 1

# Sort by document frequency
sorted_doc_freq = sorted(doc_freq.items(), key=lambda x: -x[1])

# Now pick top-N most document-frequent words
ultra_common_by_doc_freq = set([word for word, _ in sorted_doc_freq[:10]])

print(ultra_common_by_doc_freq)
```

```python
# Check how many documents "hercules" appears in
hercules_doc_count = sum("hercules" in set(doc["tokens"]) for doc in processed_corpus)
print(f"'hercules' appears in {hercules_doc_count} documents.")
```

```python
herc = "hercules"
troia = "troia"
hector = "hector"
achilles = "achilles"
iuno = "iuno"
for tragedy in processed_corpus:
    herc_count = tragedy["tokens"].count(herc)
    troia_count = tragedy["tokens"].count(troia)
    hector_count = tragedy["tokens"].count(hector)
    achilles_count = tragedy["tokens"].count(achilles)
    iuno_count = tragedy["tokens"].count(iuno)
    print(f"{tragedy['title']:<25} → 'hercules' count: {herc_count}, 'troia' count: {troia_count}, 'hector' count: {hector_count}, 'achilles' count: {achilles_count}, 'iuno' count: {iuno_count}")
```

```python
ultra_common_manual = {"hercules", "iuno", "troia", "hector", "achilles", "priamus", "danaus", "manus", "magnus", "deus", "mater", "pater", "malum", "regnum", "peto", "scelus", "dies"} #for filtered_topic_modeling_heatmap, 
#used "manus", "scelus", "magnus", "deus", "pater", "mater", "malum", "regnum", "domus", "dies", "animus", "peto" after "danaus" instead

filtered_token_lists = [
    [token for token in tragedy["tokens"] if token not in ultra_common_manual]
    for tragedy in processed_corpus
]
filtered_dictionary = corpora.Dictionary(filtered_token_lists)
filtered_corpus = [filtered_dictionary.doc2bow(filtered_tokens) for filtered_tokens in filtered_token_lists]
```

```python
filt_lda_model = LdaModel(
    corpus=filtered_corpus,
    id2word=filtered_dictionary,
    num_topics=10
)
filt_lda_model.show_topics()
```

```python
for idx, topic in filt_lda_model.print_topics(num_words=10):
    print(f"Topic #{idx + 1}: {topic}")
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.gensim_models.prepare(filt_lda_model, filtered_corpus, filtered_dictionary)
```

```python
filt_document_topics = []
for i, bow in enumerate(filtered_corpus):
    topic_dist = filt_lda_model.get_document_topics(bow)
    filt_document_topics.append((processed_corpus[i]["title"], topic_dist))
```

```python
for title, topics in filt_document_topics:
    # Get topic with highest probability
    top_topic = max(topics, key=lambda x: x[1])
    print(f"{title:<20} → Topic #{top_topic[0] + 1} (weight: {top_topic[1]:.3f})")
   
#get topic distribution for each play
for title, topics in filt_document_topics:
    print(f"\n{title}")
    for topic_id, weight in sorted(topics, key=lambda x: -x[1]):
        print(f"  Topic #{topic_id + 1}: {weight:.3f}")
```

```python
for topic_id in range(filt_lda_model.num_topics):
    topic_terms = filt_lda_model.show_topic(topic_id, topn=10)
    print(f"Gensim Topic #{topic_id + 1}: {[term for term, _ in topic_terms]}")
```

```python
rows = []
for i, (title, topics) in enumerate(filt_document_topics):
    topic_weights = [0] * filt_lda_model.num_topics
    for topic_id, weight in topics:
        topic_weights[topic_id] = weight
    rows.append([title] + topic_weights)

filt_df_topics = pd.DataFrame(rows, columns=["Title"] + [f"Topic {i+1}" for i in range(filt_lda_model.num_topics)])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(filt_df_topics.set_index("Title"), cmap="Blues", annot=True)
plt.title("Topic Distribution per Play")
plt.savefig("filtered2_topic_modeling_heatmap.png", dpi=300)
plt.show()
```

### Attempting TF-IDF before LDA

```python
from gensim import models

# Step 1: Create dictionary and BoW corpus from processed tokens
tfidf_token_lists = [tragedy["tokens"] for tragedy in processed_corpus]
dictionary = corpora.Dictionary(token_lists)
corpus = [dictionary.doc2bow(text) for text in token_lists]

# Step 2: Apply TF-IDF model to BoW corpus
tfidf_model = models.TfidfModel(corpus)
corpus_tfidf = tfidf_model[corpus]

# Step 3: Run LDA on the TF-IDF weighted corpus
lda_tfidf_model = models.LdaModel(
    corpus=corpus_tfidf,
    id2word=dictionary,
    num_topics=10,
    random_state=42,
    passes=10
)
```

```python
lda_tfidf_model.show_topics()
```

```python
for idx, topic in lda_tfidf_model.print_topics(num_words=10):
    print(f"Topic #{idx + 1}: {topic}")
```
