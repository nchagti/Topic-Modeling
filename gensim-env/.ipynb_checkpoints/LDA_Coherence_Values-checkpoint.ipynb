```python
num_topics = list(range(11)[1:])
num_keywords = 10

filtered_LDA_models = {}
filtered_LDA_topics = {}
for i in num_topics:
    filtered_LDA_models[i] = LdaModel(corpus=filtered_corpus,
                             id2word=filtered_dictionary,
                             num_topics=i,
                             chunksize=len(filtered_corpus),
                             passes=20,
                             alpha='symmetric',
                             random_state=42)

    shown_topics = filtered_LDA_models[i].show_topics(num_topics=i, 
                                             num_words=num_keywords,
                                             formatted=False) #returns a list of (topic_id, topic_words) pairs
    filtered_LDA_topics[i] = [[word[0] for word in topic[1]] for topic in shown_topics] #topic[1] is a list of (word, probability) tuples for the given topic #word[0] is the word in topic[1], word[1] would be the probability 
```

```python
#write a function to derive the Jaccard similarity of two topics

def jaccard_similarity(topic_1, topic_2):
    """
    Derives the Jaccard similarity of two topics

    Jaccard similarity:
    - A statistic used for comparing the similarity and diversity of sample sets
    - J(A,B) = (A ∩ B)/(A ∪ B)
    - where: 
    - A∩B is the number of common elements between sets
    - A∪B is the total number of unique elements in both sets.
    - In set theory, the intersection (written using the symbol "∩" between the terms) 
    - of two sets A and B, denoted by A∩B, is the set containing all elements of A that also belong to B
    - or equivalently, all elements of B that also belong to A.
    
    The value of Jaccard similarity ranges from 0 to 1:
    - J(A,B)=1 → The sets are identical.
    - J(A,B)=0 → The sets have no common elements.
    
    - Goal is low Jaccard scores for coverage of the diverse elements
    """
    intersection = set(topic_1).intersection(set(topic_2))
    union = set(topic_1).union(set(topic_2))
                    
    return float(len(intersection))/float(len(union))
```

```python
# Derive mean stability across topics by comparing the next topic model
filtered_LDA_stability = {}

for i in range(len(num_topics) - 1): #excluding the last topic count because otherwise I will run into LDA_topic[10+1] not in range error
    jaccard_sims = []
    for topic1 in filtered_LDA_topics[num_topics[i]]: #LDA_topic[1] would give one topic
        sims = []
        for topic2 in filtered_LDA_topics[num_topics[i + 1]]:
            sims.append(jaccard_similarity(topic1, topic2))  #inner loop compares topic1 with each topic in the 2-topic model; i.e. LDA_topic[4] would compare each topic against LDA_topic[5], which has 5 topics
        jaccard_sims.append(sims)
    
    filtered_LDA_stability[num_topics[i]] = jaccard_sims

filtered_mean_stabilities = [np.array(filtered_LDA_stability[i]).mean() for i in num_topics[:-1]] #list of average Jaccard similarities comparing each model with the next one — lower values mean more diversity


#another way of writing it since num_topics[0] = LDA_topics[1] 
#for i in num_topics[:-1]:  # values like 1, 2, ..., 9 #excluding the last topic count because otherwise I will run into LDA_topic[10+1] not in range error
    #jaccard_sims = []
    #for topic1 in filtered_LDA_topics[i]:
        #sims = []
        #for topic2 in filtered_LDA_topics[i + 1]:
            #sims.append(jaccard_similarity(topic1, topic2))
        #jaccard_sims.append(sims)
    
    #filtered_LDA_stability[i] = jaccard_sims
```

```python
#get topic coherences
filtered_coherences = [CoherenceModel(model=filtered_LDA_models[i], texts=filtered_token_lists, dictionary=filtered_dictionary, coherence='c_v').get_coherence()\
              for i in num_topics[:-1]]
```

```python
# derive the ideal number of topics by finding the topic count where the difference between topic coherence and topic stability is largest — i.e., where a model is both semantically coherent and sufficiently diverse
filtered_coh_sta_diffs = [                            
    filtered_coherences[i] - filtered_mean_stabilities[i]        # iterate over the indices of coherences (which we know is length 9 because mean_stabilities is length 9)
    for i in range(len(filtered_coherences))
]
filtered_coh_sta_max = max(filtered_coh_sta_diffs) # max() returns the item with the highest value # candidate for the "best" topic model: high coherence, low redundancy
filtered_coh_sta_max_idxs = [i for i, j in enumerate(filtered_coh_sta_diffs) if j == filtered_coh_sta_max] # multiple topics counts could have the same best score #collect all indices i where the score equals the maximum
filtered_ideal_topic_num_index = filtered_coh_sta_max_idxs[0] # Choose the first index among the best-scoring ones — i.e., the lowest topic number in case of ties
filtered_ideal_topic_num = num_topics[filtered_ideal_topic_num_index] # use that index to look up the actual topic count
```

```python
print(f"Ideal topic number (filtered tokens): {filtered_ideal_topic_num}")
```
