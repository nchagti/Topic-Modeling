```python
with open("../../Corpus/processed_corpus.pkl", "rb") as f:
    processed_corpus = pickle.load(f)

with open("filtered_token_lists.pkl", "rb") as f:
    filtered_token_lists = pickle.load(f)

with open('filtered_corpus.pkl', 'rb') as f:
	filtered_corpus = pickle.load(f)
filtered_dictionary = corpora.Dictionary.load('filtered_dictionary.gensim')

with open('filtered_LDA_models.pkl', 'rb') as f:
    filtered_LDA_models = pickle.load(f)

with open('filtered_LDA_topics.pkl', 'rb') as f:
    filtered_LDA_topics = pickle.load(f)

with open('filtered_coherences.pkl', 'rb') as f:
    filtered_coherences = pickle.load(f)

with open('filtered_mean_stabilities.pkl', 'rb') as f:
    filtered_mean_stabilities = pickle.load(f)

with open('filtered_coh_sta_diffs.pkl', 'rb') as f:
    filtered_coh_sta_diffs = pickle.load(f)

with open('filtered_coh_sta_max.pkl', 'rb') as f:
    filtered_coh_sta_max = pickle.load(f)

with open('filtered_coh_sta_max_idxs.pkl', 'rb') as f:
    filtered_coh_sta_max_idxs = pickle.load(f)

with open('filtered_ideal_topic_num_index.pkl', 'rb') as f:
    filtered_ideal_topic_num_index = pickle.load(f)

with open('filtered_ideal_topic_num.pkl', 'rb') as f:
    filtered_ideal_topic_num = pickle.load(f)

with open('filtered_ideal_model.pkl', 'rb') as f:
    filtered_model = pickle.load(f)
```

```python
# Get the ideal model based on earlier calculation
num_keywords = 10
filtered_model = filtered_LDA_models[filtered_ideal_topic_num]

# Print top topic words
print(f"\nTop {num_keywords} words per topic for filtered LDA({filtered_ideal_topic_num}):\n")

for topic_id, topic_words in filtered_model.show_topics(num_topics=filtered_ideal_topic_num, num_words=num_keywords, formatted=False):
    words_only = [word for word, prob in topic_words]
    print(f"Topic {topic_id + 1}: {', '.join(words_only)}")
```

```python
filtered_document_topics = []
for i, bow in enumerate(filtered_corpus):
    topic_dist = filtered_model.get_document_topics(bow)  # topic_dist is a list of tuples representing the topic distribution for the i-th document # each tuple is (topic_id, topic_weight)
    filtered_document_topics.append((processed_corpus[i]["title"], topic_dist))

for title, topics in filtered_document_topics:
    # Get topic with highest probability
    top_topic = max(topics, key=lambda x: x[1])
    print(f"{title:<20} → Filtered Topic #{top_topic[0] + 1} (weight: {top_topic[1]:.3f})")
   
#get topic distribution for each play
for title, topics in filtered_document_topics:
    print(f"\n{title}")
    top_three = sorted(topics, key=lambda x: -x[1])[:3]
    for topic_id, weight in top_three:
        print(f"  Filtered Topic #{topic_id + 1}: {weight:.3f}")
```

```python
pyLDAvis.enable_notebook()
p = pyLDAvis.gensim_models.prepare(filtered_model, filtered_corpus, filtered_dictionary, sort_topics=False)
pyLDAvis.save_html(p, '../../figures/acts_figures/acts_lda.html')
```
